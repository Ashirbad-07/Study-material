```python
#Create a 1D array of the first 10 positive integers.

import numpy as np

# Create the 1D array of the first 10 positive integers
array = np.arange(1, 11)

# Print the array
print("Array:", array)

# Find the sum of all elements in the array
sum_of_elements = np.sum(array)
print("Sum of elements:", sum_of_elements)

# Calculate the mean of the elements
mean_of_elements = np.mean(array)
print("Mean of elements:", mean_of_elements)
```

```python
#Create a 2D array of shape (3, 4) with random integers between 1 and 10.

import numpy as np

# Create the 2D array with random integers between 1 and 10
array_2d = np.random.randint(1, 11, size=(3, 4))

# Print the array
print("2D Array:\n", array_2d)

# Find the maximum value in the array
max_value = np.max(array_2d)
print("Maximum value:", max_value)

# Find the minimum value in the array
min_value = np.min(array_2d)
print("Minimum value:", min_value)

# Calculate the sum of each row
sum_of_rows = np.sum(array_2d, axis=1)
print("Sum of each row:", sum_of_rows)

# Calculate the sum of each column
sum_of_columns = np.sum(array_2d, axis=0)
print("Sum of each column:", sum_of_columns)
```

```python
#Create a 2D array of shape (3, 4) with random integers between 1 and 10.
import numpy as np
# Create the 1D array with 20 random integers between 1 and 100
array_1d = np.random.randint(1, 101, size=20)

#print the array
print("1D array:",array_1d)
# Print the elements from index 5 to 15
elements_5_to_15 = array_1d[5:16]
print("Elements from index 5 to 15:", elements_5_to_15)
# Print every second element of the array
every_second_element = array_1d[::2]
print("Every second element:", every_second_element)
```

```python
#Create a 1D array of 20 random integers between 1 and 100.

import numpy as np

# Create a 1D array of 20 random integers between 1 and 100
random_array = np.random.randint(1, 101, 20)

# Print the array
print("Array:", random_array)

# Print the elements from index 5 to 15
print("Elements from index 5 to 15:", random_array[5:16])

# Print every second element of the array
print("Every second element of the array:", random_array[::2])

```
